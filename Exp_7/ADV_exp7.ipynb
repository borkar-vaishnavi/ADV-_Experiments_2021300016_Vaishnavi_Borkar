```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('loan_new.csv')

# Set the color palette for seaborn
sns.set_palette("bright")

# Bar Chart: Number of Loans by Loan Status
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Loan Status')
plt.title('Number of Loans by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Pie Chart: Distribution of Home Ownership
home_ownership_counts = data['Home Ownership'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(home_ownership_counts, labels=home_ownership_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title('Distribution of Home Ownership')
plt.show()

# Scatter Plot: Annual Income vs Current Loan Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Annual Income', y='Current Loan Amount', hue='Loan Status', palette='Set1')
plt.title('Annual Income vs Current Loan Amount')
plt.xlabel('Annual Income')
plt.ylabel('Current Loan Amount')
plt.show()

# Histogram: Credit Score Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Credit Score'], bins=20, kde=True, color='purple')
plt.title('Credit Score Distribution')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.show()

# Box Plot: Distribution of Credit Scores by Loan Status
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Loan Status', y='Credit Score', palette='Set2')
plt.title('Credit Score Distribution by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Credit Score')
plt.xticks(rotation=45)
plt.show()

# Violin Plot: Distribution of Annual Income by Loan Term
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='Term', y='Annual Income', palette='Set3')
plt.title('Annual Income Distribution by Loan Term')
plt.xlabel('Loan Term')
plt.ylabel('Annual Income')
plt.xticks(rotation=45)
plt.show()
```
