```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv("grocery.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
plt.figure(figsize=(9, 6))  # Adjust the width and height as needed
sns.barplot(data, x='Category', y='Sales')
plt.show()
#inference = Grain and Meat-Fish category items were bought frequently in TamilNadu
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8, 8))  # Adjust the size as needed
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Category')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data ,x = "Category",bins=5)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10, 8))  # Adjust the width and height as needed
sns.barplot(data, x='City', y='Sales')
plt.xticks(rotation=45)
#highest sales in Vellore
```

```python
category_sales = data.groupby('City')['Sales'].sum()

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust the size as needed
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by City')
plt.show()
```

```python
sns.scatterplot(data, x='Category',y='Region')
plt.xticks(rotation=45)
```

```python
import pandas as pd

# Assuming df is your DataFrame
data['Order_Date'] = pd.to_datetime(data['Order_Date'], dayfirst=True, errors='coerce')
```

```python
data.Order_Date.head()
```

```python
# Resampling sales data by month
monthly_sales = data.resample('M', on='Order_Date').sum()
monthly_sales_sales=monthly_sales['Sales']
monthly_sales_sales.head()
```

```python
import matplotlib.pyplot as plt

# Plotting the monthly sales
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales_sales, label='Monthly Sales', color='blue')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
  # Scatter plot with Sales on the x-axis and Profit on the y-axis
  plt.figure(figsize=(10, 6))
  plt.scatter(monthly_sales['Sales'], monthly_sales['Profit'], s=monthly_sales['Discount'], c='blue', alpha=0.5)
  plt.xlabel('Sales')
  plt.ylabel('Profit')
  plt.title('Scatter Plot of Sales vs. Profit')
  plt.grid(True)
  plt.show()
```
